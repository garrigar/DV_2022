## Тестовый датафрейм "mpg"

Категории датафрейма mpg:

- manufacturer – производитель,
- model – модель,
- displ – рабочий объём двигателя в литрах,
- year – год выпуска,
- cyl – число цилиндров,
- trans – тип коробки передач,
- drv – тип привода,
- cty – расход топлива в городе (миль/галлон),
- hwy – расход топлива на трассе (миль/галлон)

### запустить без "group=1"? Почему?

Ответ: без ```group=1``` geom_bar посчитает количество вхождений данных по отдельности, а не относительно друг друга.

````python
# Type your code here...
ggplot(data=diamonds) +\
geom_bar(mapping=aes(x="cut", y="..prop.."), group=1)
````

### Зачем тут coord_fixed()? Что делает geom_abline()?

- coord_fixed - сопоставляет размеры осей, чтобы один не был растянутым по сравнению с другим.
- geom_abline - создаёт линию по наклону и пересечению (здесь прямая y = x)

````python
ggplot(data=mpg, mapping=aes(x="cty", y="hwy")) +\
geom_point() +\
geom_abline() +\
coord_fixed()
````
