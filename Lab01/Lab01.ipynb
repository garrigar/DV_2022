```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# plt.rcParams["animation.html"] = "html5"
```

History of Rock

This dataset contains 5484 Rock songs including their Spotify audio features. This dataset spans over 70 years from when Rock was invented in 1950, to today.

```python
df = pd.read_csv('../data/history-of-rock-spotify.csv').drop(columns='index')
df
```

```python
df.dtypes
```

Song popularities by key

Bad view

```python
df.plot(x='key', y='popularity', kind='scatter')
```

Good view

```python
df.plot.box(column=['popularity'], by='key')
```

for each release date, average popularity per different musical key

```python
df0 = df.pivot_table(index='release_date', columns='key', values='popularity')
df0
# for each release date, average popularity per different musical key
```

Bad view

```python
X, Y = np.meshgrid(df0.index, df0.columns)
Z = df0.transpose().values

fig = plt.figure(figsize=(16, 16))
ax = fig.add_subplot(projection='3d')
surf = ax.plot_wireframe(X, Y, Z)

ax.set_xlabel('year')
ax.set_ylabel('key')
ax.set_zlabel('popularity')

plt.show()
```

better view

```python
fig = plt.figure(figsize=(16, 16))
ax = fig.add_subplot(projection='3d')

df1 = df0.fillna(0)

key_ticks = list(range(len(df1.columns)))
# key_ticks = list(range(len(df1.index)))

for k in key_ticks:
    ser = df1[df1.columns[k]]
    # ser = df1.iloc[k]
    xs = ser.index.tolist()
    ys = ser.values.tolist()
    # print(len(xs), len(ys))
    ax.bar(xs, ys, zs=k, zdir='y', alpha=0.5)

ax.set_xlabel('year')
ax.set_ylabel('key')
ax.set_zlabel('popularity')

ax.set_yticks(key_ticks)

plt.show()
```

Good view

```python
from matplotlib.animation import FuncAnimation
```

```python
from IPython.display import HTML
```

```python
fig, ax = plt.subplots()

line, = ax.plot([], linestyle='', marker='+')
# title = ax.set_title('asdf')

title = ax.text(0.5, 1.05, "", # bbox={'facecolor':'w', 'alpha':0.5, 'pad':5},
                transform=ax.transAxes, ha="center")

ax.set_xlim(df0.index[0], df0.index[-1])
ax.set_ylim(0, 110)
# ax.legend()

x = df0.index.tolist()

def animate(frame):
    key = df0.columns[frame]
    y = df0[key].values.tolist()
    
    line.set_data((x, y))
    title.set_text(f'Average popularity of songs of key = {key} by year')
    # line.set_label(f'{key}')
    return line

ax.grid()
    
anim = FuncAnimation(fig, animate, frames=12, interval=1000)

HTML(anim.to_html5_video())
```
